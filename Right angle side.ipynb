# Find the third side of Right Angle Triangle

```python
def pythagorus(base, height , hypotenuse):
    if base==str("x"):
        return ("base =" , str((hypotenuse**2-height**2)**0.5))
    elif height==str("x"):
        return("height =",str((hypotenuse**2-base**2)**0.5))
    elif hypotenuse==str("x"):
        return("hypotenuse :",str((base**2+height**2)))
    
    
print(pythagorus(4,5,'x'))
print(pythagorus(4,'x',5))
print(pythagorus('x',4,5))
```
