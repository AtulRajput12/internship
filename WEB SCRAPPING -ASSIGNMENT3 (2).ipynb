# 2. In the above question, now scrape the following details of each product

 Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
urls=[]

for i in driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]'):
    urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
urls
```

```python
#defining variables
product_name=[]
expected_delivery=[]
price=[]
availability=[]
exchange=[]
brand=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    #fetching brand name
    try:
        Brand=driver.find_element_by_xpath('//span[@class="a-size-base"][14]')
        brand.append(Brand.text.replace("\n","new line"))
        
    except:
        brand.append('-')
    
    #fetching product name
    try:
        product=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
        product_name.append(product.text.replace("\n","new line"))
        
    except:
        product_name.append('-')
    

    #fetching price
    try:
        Price=driver.find_element_by_xpath('//span[@class="a-price aok-align-center"][1]')
        price.append(Price.text.replace("\n","new line"))
        
    except:
        price.append('-')
        
        
    #fetching availability
    try:
        avail=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        availability.append(avail.text.replace("\n","new line"))
        
    except:
        availability.append('-')
        
        
    #fetching exchange
    try:
        change=driver.find_element_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')
        exchange.append(change.text.replace("\n","new line"))
        
    except:
        exchange.append('-')
        
    
    #fetching expected delivery
    try:
        delivery=driver.find_element_by_id('ddmDeliveryMessage')
        expected_delivery.append(delivery.text.replace("\n","new line"))
        
    except:
        expected_delivery.append('-')
```

```python
print("product-",len(product_name))
print("price-",len(price))
print("availabilty-",len(availability))
print("exchange-",len(exchange))
print("expected delivery-",len(expected_delivery))
print("brand-",len(brand))
```

```python
data=pd.DataFrame()
data['Product_name']=product_name
data['Price']=price
data['Availability']=availability
data['Expected_delivery']=expected_delivery
data['Exchange']=exchange
data['Product_url']=urls
data
```
